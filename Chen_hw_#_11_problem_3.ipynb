```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f_int(x, t, a,b,c):
    '''Defines the function'''
    return [-x[1]-x[2], x[0]+a*x[1], b+x[2]*(x[0]-c)]
```

```python
x_0 = np.array([0,0,0])
t=np.linspace(0.0, 3200.0,320000)
x_sol = odeint(f_int, x_0, t, args=(0.1,0.1,4))
ax = plt.axes(projection="3d") # Data for a three-dimensional line
ax.plot3D(x_sol[:,0], x_sol[:,1], x_sol[:,2], 'gray')
plt.show()
```

It is a closed orbital. Finally, it will repeat the same trajectory. 

```python
x_0 = np.array([0,0,0])
t=np.linspace(0.0, 3200.0,320000)
x_sol = odeint(f_int, x_0, t, args=(0.1,0.1,8))
ax = plt.axes(projection="3d") # Data for a three-dimensional line
ax.plot3D(x_sol[:,0], x_sol[:,1], x_sol[:,2], 'gray')
plt.show()
```

When $c=8$, the final trojectory is quite different from one above. In this case, the trajectory is splitted into outer one and inner one. After the long time, the particle will repeat this trajectory again and again.

```python
x_0 = np.array([0,0,0])
t=np.linspace(0.0, 6400.0,640000)
x_sol = odeint(f_int, x_0, t, args=(0.1,0.1,16))
ax = plt.axes(projection="3d") # Data for a three-dimensional line
ax.plot3D(x_sol[:,0], x_sol[:,1], x_sol[:,2], 'gray')
plt.show()
```

When $c=16$, the trajectory is chaotic. We cannot find the rule of the time evolution of trajectory.
