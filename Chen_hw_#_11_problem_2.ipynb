```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits import mplot3d
import time
```

```python
def liss(t, a, b, amp_a, amp_b, delt):
    """returns the Lissajous figure for the given paramters
    """
    return np.array([amp_a*np.sin(a*t+ delt), amp_b*np.sin(b*t)])
```

```python
a= 1.0
amp_a = 1.0
b = a*np.e
amp_b = amp_a
temp = np.linspace(0.0, 2*np.pi*800, 40000)
start = time.time()
Xx, Yy = liss(temp, a, b, amp_a, amp_b,0)
XXx=(Xx+1)*100
YYy=(Yy+1)*100

XX=np.zeros(40000,dtype=int)
YY=np.zeros(40000,dtype=int)

for i in range(40000):
    XX[i]=np.int(XXx[i])
    YY[i]=np.int(YYy[i])

Nx=np.bincount(XX)
Ny=np.bincount(YY)
z=np.zeros((200,200))
Nz=np.zeros((200,200))
print(XX)
print(YY)
```

```python
for i in range(200):
    for j in range(200):
        z[i,j]=Nx[i]*Ny[j]/40000/40000
for k in range(40000):
    Nz[XX[k],YY[k]]=Nz[XX[k],YY[k]]+1/40000
x=np.linspace(-1,1,200)
y=np.linspace(-1,1,200)
X,Y=np.meshgrid(x, y)
print(np.shape(X))
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface')
plt.show()
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Nz, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface')
plt.show()
```
