$L=\frac{1}{2}\mu\left(\dot{x}^{2}+\dot{y}^{2}+\dot{z}^{2}\right)+\frac{k}{\sqrt{x^{2}+y^{2}+z^{2}}}+\mu gz$

The new coordinate is

$x=\sqrt{uv}\cos\left(\phi\right)$

$y=\sqrt{uv}\sin\left(\phi\right)$

$z=\frac{1}{2}\left(u-v\right)$

the unit vectors are

$\hat{u}=\sqrt{\frac{v}{u+v}}\cos\left(\phi\right)\hat{x}+\sqrt{\frac{v}{u+v}}\sin\left(\phi\right)\hat{y}+\sqrt{\frac{u}{u+v}}\hat{z}$

$\hat{v}=-\sqrt{\frac{u}{u+v}}\cos\left(\phi\right)\hat{x}-\sqrt{\frac{u}{u+v}}\sin\left(\phi\right)\hat{y}+\sqrt{\frac{v}{u+v}}\hat{z}$

$\hat{\phi}=-\sin\left(\phi\right)\hat{x}+\cos\left(\phi\right)\hat{y}$

$\dot{\hat{u}}=\frac{1}{2v\left(u+v\right)}\left(\dot{u}v-\dot{v}u\right)\hat{v}-\dot{\phi}\sqrt{\frac{v}{u+v}}\hat{\phi}$

$\dot{\hat{v}}=\frac{1}{2u\left(u+v\right)}\left(\dot{v}u-\dot{u}v\right)\hat{u}+\dot{\phi}\sqrt{\frac{u}{u+v}}\hat{\phi}$

$\dot{\hat{\phi}}=\dot{\phi}\sqrt{\frac{u}{u+v}}\hat{u}-\dot{\phi}\sqrt{\frac{u}{u+v}}\hat{v}$

The vector of the position can be expressed as

$\vec{r}=x\hat{x}+y\hat{y}+z\hat{z}$

$\vec{r}=\sqrt{uv}\cos\left(\phi\right)\hat{x}+\sqrt{uv}\sin\left(\phi\right)\hat{y}+\frac{1}{2}\left(u-v\right)\hat{z}$

$\vec{r}=\frac{\sqrt{u\left(u+v\right)}}{2}\hat{u}-\frac{\sqrt{v\left(u+v\right)}}{2}\hat{v}$

Differentiated by time, it can be the velocity

$\dot{\vec{r}}=\frac{1}{2}\sqrt{\frac{u+v}{u}}\dot{u}\hat{u}-\frac{1}{2}\sqrt{\frac{u+v}{v}}\dot{v}-\sqrt{uv}\dot{\phi}\hat{\phi}$

Therefore, the Lagrange can be rewritten as 

$L=\frac{1}{8}\mu\left(\frac{u+v}{u}\right)\dot{u}^{2}+\frac{1}{8}\mu\left(\frac{u+v}{v}\right)\dot{v}^{2}+\frac{1}{2}\mu uv\dot{\phi}^{2}+\frac{k}{u+v}+\frac{1}{2}\mu g\left(u-v\right)$

The momentum can be written down as

$\frac{\partial L}{\partial\dot{u}}=P_{u}=\frac{1}{4}\mu\left(\frac{u+v}{u}\right)\dot{u}$

$\frac{\partial L}{\partial\dot{v}}=P_{v}=\frac{1}{4}\mu\left(\frac{u+v}{v}\right)\dot{v}$

$\frac{\partial L}{\partial\dot{\phi}}=P_{\phi}=\mu uv\dot{\phi}$

Therefore, the Hamiltonian can be expressed as

$H=\frac{2up_{u}^{2}}{\mu\left(u+v\right)}+\frac{2vp_{v}^{2}}{\mu\left(u+v\right)}+\frac{p_{\phi}^{2}}{2\mu uv}-\frac{k}{u+v}-\frac{1}{2}\mu g\left(u-v\right)$

Because Hamitonian is time independent. The total energy is conserved. Let $H(u,v,\phi)=E$

$E=\frac{2up_{u}^{2}}{\mu\left(u+v\right)}+\frac{2vp_{v}^{2}}{\mu\left(u+v\right)}+\frac{p_{\phi}^{2}}{2\mu\left(u+v\right)}\left(\frac{1}{u}+\frac{1}{v}\right)-\frac{k}{u+v}-\frac{1}{2}\mu g\left(u-v\right)$

$\mu\left(u+v\right)E=2up_{u}^{2}+2vp_{v}^{2}+\frac{p_{\phi}^{2}}{2}\left(\frac{1}{u}+\frac{1}{v}\right)-k-\frac{1}{2}\mu g\left(u^{2}-v^{2}\right)$

The equation of the characteristic function for $\phi$ is

$\frac{\partial W_{\phi}}{\partial\phi}=\alpha_{\phi}$

$W_{\phi}=\alpha_{\phi}\phi$

Therefore, we can get the equation 

$\mu\left(u+v\right)E=2up_{u}^{2}+2vp_{v}^{2}+\frac{\alpha_{\phi}^{2}}{2}\left(\frac{1}{u}+\frac{1}{v}\right)-k-\frac{1}{2}\mu g\left(u^{2}-v^{2}\right)$

The equation can be separated as

$2up_{u}^{2}-\frac{1}{2}\mu gu^{2}-\frac{k}{2}=\mu uE-\frac{\alpha_{\phi}^{2}}{2u}$

$2vp_{v}^{2}+\frac{1}{2}\mu gv^{2}-\frac{k}{2}=\mu vE-\frac{\alpha_{\phi}^{2}}{2v}$

Hence, the charactistic function for $u$ and $v$ respectively are

$\left(\frac{\partial W_{u}}{\partial u}\right)^{2}=\mu E+\frac{1}{4}\mu gu+\frac{k}{4u}-\frac{\alpha_{\phi}^{2}}{2u^{2}}$

$\left(\frac{\partial W_{v}}{\partial v}\right)^{2}=\mu E-\frac{1}{4}\mu gv+\frac{k}{4v}-\frac{\alpha_{\phi}^{2}}{2v^{2}}$

The solution will be

$W_{u}=\int\sqrt{\mu E+\frac{1}{4}\mu gu+\frac{k}{4u}-\frac{\alpha_{\phi}^{2}}{2u^{2}}}du$

$W_{v}=\int\sqrt{\mu E-\frac{1}{4}\mu gv+\frac{k}{4v}-\frac{\alpha_{\phi}^{2}}{2v^{2}}}dv$
